```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.spatial import Delaunay
```

```python
# Read data

points = np.genfromtxt('data.txt',delimiter=',')
```

```python
points
```

```python
plt.scatter(points[:,0],points[:,1])
```

```python
tri = Delaunay(points)
```

```python
plt.triplot(points[:,0],points[:,1],tri.simplices.copy())

plt.show()
```

```python
tri=Delaunay(points[0:38,:])
plt.triplot(points[0:38,0],points[0:38,1],tri.simplices.copy())
plt.show()
```
